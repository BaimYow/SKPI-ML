```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset from url
url ='https://raw.githubusercontent.com/diazzzz/Buku-Machine-Learning/refs/heads/main/data_car.csv'
df = pd.read_csv(url)

#menampilkan  5 data teratas
df.head()
```

```python
df.count()
```

```python
#tampilkan fitur data seluruh fitur
df.dtypes
```

```python
df.shape
```

```python
#backup data ke df_orginal
df_original = df.copy()

#menghitung data duplikat menjadi variabel duplikat_row_df
duplikat_rows_df = df[df.duplicated()]
print("jumlah data duplikat", duplikat_rows_df.shape)
```

```python
df.drop_duplicates(inplace=True)
```

```python
#drop data duplikat
df = df.drop_duplicates()
df.shape
```

```python
#menampilkan data missing value
df.isnull().sum()
```

```python
#menghapus data yang missing value
df = df.dropna()
df.shape
```

```python
#memilih fitur
print(df.columns)

df = df.drop(['Engine Fuel Type','Market Category','Number of Doors','Engine HP'], axis=1)
df.head()
```

```python
#membuat histogram dari feature merek
df['Make'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Jumlah Kendaraan Berdasarkan Merk')
plt.ylabel('Jumlah Kendaraan')
plt.xlabel('Merk')
```
